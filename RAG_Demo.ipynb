```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("/content/1706.03762v7 (1).pdf")
```

```python
pages = loader.load_and_split()
```

```python
pages[0].page_content
```

```python
pip install langchain-text-splitters
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
chunks = text_splitter.split_documents(pages)
```

```python
len(chunks)
```

```python
pip install langchain_huggingface
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
hf_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
pip install langchain_chroma
```

```python
from langchain_chroma import Chroma
db = Chroma.from_documents(chunks, hf_embeddings)
```

```python
reteriver = db.as_retriever()
```

```python
user_input= input("enter your user_query: ")
```

```python
relevants_chunks = reteriver.invoke(user_input)
```

```python
final_context = "\n".join([chunk.page_content  for chunk in relevants_chunks])
```

```python
final_context
```

```python
prompt = f'''
you are expert in analzing user query. refer below user query and context to build response.
if context is not aligned with user query , reply with 'no context found'.
Don't use knowledge to build response.

user query : {user_input}

context : {final_context}
'''
```

```python
print(prompt)
```
